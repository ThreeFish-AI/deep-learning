# 激活函数

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

## 激活函数图形绘制

```python
def draw(X, activation_func, title, color='orange'):
    """
    绘制激活函数图形
    :param X: 输入数据
    :param activation_func: 激活函数
    :param title: 图形标题
    :param color: 图形颜色
    """

    # 创建一个新的图形，figsize 参数指定图形的大小为 6 x 3
    plt.figure(figsize=(6, 3))

    Y = activation_func(X)
    plt.plot(X, Y, label=title, color=color)
    plt.title(title)
    plt.grid()
    plt.xlabel('Input')
    plt.ylabel('Output')
```

```python
# 创建一个从 -10 到 10 的线性空间，共 400 个数据点
X = np.linspace(-10, 10, 400)
```

```python
def step_single(x):
    """阶跃函数（单个数据计算）"""
    return 1 if x > 0 else 0

def step(x):
    """阶跃函数（批量数据计算）"""
    return np.array(x > 0, dtype=np.int32)
```

```python
draw(X, step, 'Step Function')
```

```python
def sigmoid_single(x):
    """Sigmoid 函数（单个数据计算）"""
    return 1 / (1 + math.exp(-x))

def sigmoid(x):
    """Sigmoid 函数（批量数据计算）"""
    return 1 / (1 + np.exp(-x))
```

```python
draw(X, sigmoid, 'Sigmoid Function')
```

```python
def relu_single(x):
    """ReLU 函数（单个数据计算）"""
    return max(0, x)

def relu(x):
    """ReLU 函数（批量数据计算）"""
    return np.maximum(0, x)
```

```python
draw(X, relu, 'Relu Function')
```

```python
def leaky_relu_single(x, alpha=0.01):
    """Leaky ReLU 函数（单个数据计算）"""
    return x if x > 0 else alpha * x

# 4. Leaky ReLU
def leaky_relu(x, alpha=0.01):
    """Leaky ReLU 函数（批量数据计算）"""
    return np.where(x > 0, x, alpha * x)
```

```python
draw(X, leaky_relu, 'Leaky ReLU Function')
```

```python
def softmax_single(x):
    """Softmax 函数（单个数据计算）"""
    e_x = np.exp(x - np.max(x))  # 减去最大值以提高数值稳定性
    return e_x / e_x.sum()

def softmax(x):
    """Softmax 函数（批量数据计算）"""
    e_x = np.exp(x - np.max(x))  # 减去最大值以提高数值稳定性
    return e_x / e_x.sum(axis=0)
```

```python
draw(X, softmax, 'Softmax Function')
```

```python
def tanh_single(x):
    """Tanh 函数（单个数据计算）"""
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))

def tanh(x):
    """Tanh 函数（批量数据计算）"""
    return np.tanh(x)
```

```python
draw(X, tanh, 'Tanh Function')
```

```python
def identity(x):
    """恒等函数"""
    return x
```

```python
draw(X, identity, 'Identity Function')
```
